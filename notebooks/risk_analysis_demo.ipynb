```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a dummy CSV file if it doesn't exist
try:
    df = pd.read_csv('sample_health_data.csv')
except FileNotFoundError:
    data = {
        'Patient_ID': [1, 2, 3, 4, 5],
        'Age': [30, 45, 22, 60, 35],
        'Weight_kg': [70, 85, 55, 90, 65],
        'Height_cm': [170, 180, 160, 175, 168],
        'Blood_Pressure_Systolic': [120, 130, 110, 140, 125],
        'Blood_Pressure_Diastolic': [80, 85, 70, 90, 80]
    }
    df = pd.DataFrame(data)
    df.to_csv('sample_health_data.csv', index=False)
    print("Created a dummy 'sample_health_data.csv' file.")

# Display dataset
df.head()
```

```python
# Simple risk categorization logic
def classify_risk(score):
    if score >= 80:
        return "High"
    elif score >= 50:
        return "Medium"
    else:
        return "Low"

df["predicted_risk"] = df["risk_score"].apply(classify_risk)

# Count risk levels
df["predicted_risk"].value_counts()
```

```python
# Plot risk distribution
df["predicted_risk"].value_counts().plot(kind='bar')
plt.title("Malnutrition Risk Distribution")
plt.xlabel("Risk Level")
plt.ylabel("Number of Individuals")
plt.show()
```
